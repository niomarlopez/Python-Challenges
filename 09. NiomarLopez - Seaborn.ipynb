```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
penguin_df = sns.load_dataset("penguins")
penguin_df.head()
```

# Scatter Plots

Create a scatter plot that includes the following parameters:
- bill_length_mm on the x-axis
- bill_depth_mm on the y-axis
- use hue to indicate species

```python
sns.scatterplot(data = penguin_df, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species').set(title = 'Bill Length vs Bill Depth')
```

# KDE Plot

Create a kde plot that includes the following parameters:
- display the distribution of body_mass_g 
- use hue to indicate 'island'

```python
sns.kdeplot(data = penguin_df , x = 'body_mass_g', hue = 'island').set(title = 'The chances of finding a penguin with x-bodymass on an island')
```

# Correlation Matrix + Heat Map

Create a correlation matrix for the dataframe.  Use a heatmap to show the strength of the correlations.  

```python
correlation = penguin_df.corr(numeric_only = True)
correlation
```

```python
sns.heatmap(data=correlation, annot=True).set(title = 'Correlation between variables')
```

# Pairplot

Create a pairplot for the DataFrame.   
- What (if any) variables/columns are positively correlated? 
- What (if any) variables/columns are negatively correlated? 
- What (if any) variables/columns are not correlated at all?  

```python
sns.pairplot(data = penguin_df, hue = 'species', corner = True)
```

## Positive Correlations

- `'flipper_length_mm'` to `'body_mass_g'`
- `'bill_length_mm'` to `['body_mass_g', 'flipper_length_mm']`

## Negative Correlations

- `'bill_length_mm'` to - `'bill_depth_mm'`
- `'bill_depth_mm'` to `['body_mass_g', 'flipper_length_mm']`

# Swarm Plot

Create a swarm plot with the following features:
- 'island' on the x-axis.
- 'flipper length' on the y-axis.
- hue to indicate 'species'.

```python
sns.swarmplot(data = penguin_df, x='island', y= 'flipper_length_mm', hue='species')
```
